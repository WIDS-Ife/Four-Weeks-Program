```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
iris = pd.read_csv('iris.csv')
wine = pd.read_csv('wine.csv')
```

# visualizing iris dataset

```python

iris.head()
```

```python
#figure and axis
fig, ax = plt.subplots()

ax.scatter(iris['sepal_length'], iris['sepal_width'], label = 'sepal')
ax.scatter(iris['petal_length'], iris['petal_width'], label = 'petal')
ax.set_title('iris dataset')
ax.set_xlabel('length')
ax.set_ylabel('width')
ax.legend()    
```

```python
x_data = range(0, iris.shape[0])
col = iris.columns

fig, ax = plt.subplots()
#PLOT EACH COLUMN

for i in col:
    ax.plot(x_data, iris[i], label= i)
ax.set_title('Iris Dataset')
ax.legend()
```

```python
import numpy as np
np.unique(iris.species)
```

```python
colors = {'setosa':'r', 'versicolor':'g', 'virginica':'b'}
# create a figure and axis
fig, ax = plt.subplots()
# plot each data-point
for i in range(len(iris['sepal_length'])):
    ax.scatter(iris['sepal_length'][i], iris['sepal_width'][i],color=colors[iris['species'][i]])
# set a title and labels
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')

```
